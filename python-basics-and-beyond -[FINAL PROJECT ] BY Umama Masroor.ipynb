# COMSATS University Course Management System

**Scenario Overview**

Imagine you're a student at COMSATS University, managing your various courses throughout a semester. Each course, like Programming Fundamentals, OOPS, Python, Statistics, and Calculus, has its own specific characteristics and requirements. However, they also share common features.

# 1. Base Class: Course

```python
class Course:
    def __init__(self, course_name, credits, instructor):
        self.__course_name = course_name  # Course name
        self.__credits = credits            # Credits for the course
        self.__instructor = instructor      # Instructor name

    def get_course_info(self):             # Get course details
        return f"Course: {self.__course_name}, Credits: {self.__credits}, Instructor: {self.__instructor}"

    def get_syllabus(self):                 # Get the syllabus
        return f"Syllabus for {self.__course_name}: [Details of the syllabus]"

    def enroll_student(self, student_name): # Enroll a student in the course
        return f"{student_name} has been enrolled in {self.__course_name}."
```

# 2. Derived Classes: ProgrammingCourse and MathematicsCourse
# Inheritance
Inheritance allows courses to share common features while also having specific characteristics.

**Base Class: Course**

**Common attributes:** course_name, credits, instructor

**Common methods:** get_syllabus(), enroll_student()

**Derived Classes:**
ProgrammingCourse (inherits from Course):

**Additional attribute:** programming_language

**Method:** get_practical_exercises()
MathematicsCourse (inherits from Course):

**Additional attribute:** math_field (e.g., Calculus, Statistics)
Method: get_formula_sheet()
1.  **Programming Course**
2.  **Mathematics Course**










```python
class ProgrammingCourse(Course):
    def __init__(self, course_name, credits, instructor, programming_language):
        super().__init__(course_name, credits, instructor)  # Get details from Course
        self.programming_language = programming_language      # Specific to programming courses

    def get_practical_exercises(self):  # Get exercises for programming
        return f"Practical exercises for {self.programming_language} in {self.get_course_info()}."

class MathematicsCourse(Course):
    def __init__(self, course_name, credits, instructor, math_field):
        super().__init__(course_name, credits, instructor)  # Get details from Course
        self.math_field = math_field                            # Specific to math courses

    def get_formula_sheet(self):  # Get formula sheet for math
        return f"Formula sheet for {self.math_field} in {self.get_course_info()}."
```

# Polymorphism
Polymorphism allows different types of courses to have a method that behaves differently based on the course type.

Both ProgrammingCourse and MathematicsCourse can have a method called get_schedule(). However, the output will be different:

**ProgrammingCourse:** Might include coding assignments and lab sessions.
**MathematicsCourse:** Might include theory lectures and problem-solving sessions.

```python
class ProgrammingCourse(Course):
    # Other methods...
    def get_schedule(self):  # Schedule for programming courses
        return f"Schedule: Coding assignments and lab sessions for {self.get_course_info()}."

class MathematicsCourse(Course):
    # Other methods...
    def get_schedule(self):  # Schedule for math courses
        return f"Schedule: Theory lectures and problem-solving sessions for {self.get_course_info()}."
```

# Encapsulation
**Private Attributes:** These are the details of a class (like course information) that you dont want everyone to change directly. You hide them by using a double underscore (__).

**Example:** __course_name, __credits, and __instructor in our Course class.

**Public Methods:** These are the functions that allow you to access or change the private attributes safely. Think of them as the "safe doors" to get in and out of the private data.

Example:  Methods like get_course_name(), get_credits(), and enroll_student() let you interact with the course details without exposing them directly.

```python
class Course:
    def __init__(self, course_name, credits, instructor):
        self.__course_name = course_name  # Private attribute
        self.__credits = credits            # Private attribute
        self.__instructor = instructor      # Private attribute

    def get_course_info(self):             # Public method to get details
        return f"Course: {self.__course_name}, Credits: {self.__credits}, Instructor: {self.__instructor}"

```

# Abstraction
Abstraction allows you to interact with a course without needing to understand all the underlying details.

You can enroll in a course by simply calling the enroll_student() method without needing to know how the enrollment process works behind the scenes (like database operations, checks, etc.).

```python
class Course:
    # Other methods...
    def enroll_student(self, student_name): # Enroll a student in the course
        return f"{student_name} has been enrolled in {self.__course_name}."
```

# Example Usage 

```python
class Course:
    def __init__(self, course_name: str, credits: int, instructor: str):
        self.__course_name = course_name  # Private attribute
        self.__credits = credits            # Private attribute
        self.__instructor = instructor      # Private attribute

    def get_course_name(self) -> str:
        """Returns the name of the course."""
        return self.__course_name

    def get_credits(self) -> int:
        """Returns the number of credits for the course."""
        return self.__credits

    def get_instructor(self) -> str:
        """Returns the name of the instructor."""
        return self.__instructor

    def get_syllabus(self) -> str:
        """Returns the syllabus for the course."""
        return f"Syllabus for {self.__course_name}"

    def enroll_student(self, student_name: str) -> str:
        """Enrolls a student in the course."""
        if not student_name:
            raise ValueError("Student name cannot be empty.")
        return f"{student_name} has been enrolled in {self.__course_name}."


class ProgrammingCourse(Course):
    def __init__(self, course_name: str, credits: int, instructor: str, programming_language: str):
        super().__init__(course_name, credits, instructor)
        self.programming_language = programming_language  # Additional attribute

    def get_practical_exercises(self) -> str:
        """Returns practical exercises for the programming course."""
        return f"Practical exercises for {self.get_course_name()} in {self.programming_language}."

    def get_schedule(self) -> str:
        """Returns the schedule for the programming course."""
        return f"Schedule for {self.get_course_name()}: Coding assignments and lab sessions."

    def info(self) -> str:
        """Returns course information."""
        return f"Course Name: {self.get_course_name()}, Credits: {self.get_credits()}, Instructor: {self.get_instructor()}"


class MathematicsCourse(Course):
    def __init__(self, course_name: str, credits: int, instructor: str, math_field: str):
        super().__init__(course_name, credits, instructor)
        self.math_field = math_field  # Additional attribute

    def get_formula_sheet(self) -> str:
        """Returns the formula sheet for the mathematics course."""
        return f"Formula sheet for {self.get_course_name()} in {self.math_field}."

    def get_schedule(self) -> str:
        """Returns the schedule for the mathematics course."""
        return f"Schedule for {self.get_course_name()}: Theory lectures and problem-solving sessions."

    def info(self) -> str:
        """Returns course information."""
        return f"Course Name: {self.get_course_name()}, Credits: {self.get_credits()}, Instructor: {self.get_instructor()}"


# Create instances of ProgrammingCourse and MathematicsCourse
prog_course = ProgrammingCourse("Python Programming", 1, "Hania", "Python")
math_course = MathematicsCourse("Calculus", 2, "Aliyan", "Calculus")

# Display course information
print(prog_course.info())
print(math_course.info())

# Display course schedules
print(prog_course.get_schedule())
print(math_course.get_schedule())

# Enroll students
print(prog_course.enroll_student("IMAMA"))
print(math_course.enroll_student("BARIK"))

# Display practical exercises and formula sheets
print(prog_course.get_practical_exercises())
print(math_course.get_formula_sheet())
```
