**Question 1:**

Class Attendance using a List
You are keeping track of student attendance using a list. Sometimes students may enter their names more than once.

**Step 1: Create a list called attendance_list with the names:**

["John", "Emma", "John", "Ava", "Sophia", "Emma"]

**Step 2: Print the list to see the attendance.**

**Step 3: Count how many names (including duplicates) are in the list and print the total number.**


image.png

```python
# 1: Creating a list of student names for attendance check:
Attendance_List = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# 2: Print the list to see the attendance:
print("Attendance_List:", Attendance_List)

# 3: Count how many names (including duplicates) are in the list:
total_names = len(Attendance_List)  # Corrected capitalization of the variable
print("Total number of names (including duplicates):", total_names)
```

**Question 2:**

 Unique Attendance using a Set
To find out how many unique students attended the class, use a set. A set automatically removes duplicates.

**Step 1: Convert the attendance_list from Question 1 into a set.**

**Step 2: Print the set to see the unique student names.**

**Step 3: Count and print the total number of unique students.**


image.png

```python
# Step 1: Defining the attendance list with duplicates
Attendance_List = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# Step 2: Converting the list to a unique student names.
Unique_Attendance = set(Attendance_List)

# Step 3: Printing the set to see the unique student names
print("Unique Attendance Set:",Unique_Attendance)

# Step 4: Counting and printing the total number of unique students
total_names = len(Unique_Attendance)
print("Total Unique Attendance:", total_names)
```

**Question 4:**

 Count Attendance using a Dictionary
You want to count how many times each student attended the class.

**Step 1: Using the attendance_list from Question 1, create a dictionary where each student's name is a key, and the number of times they attended is the value.**

**Step 2: Print the dictionary to see each student's attendance count.**


image.png

```python
# Step 1: Using the attendance_list from Question 1
attendance_list = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# Step 2: Creating an empty dictionary to store
attendance_count = {}

# Step 3: Counting the attendance
for student in attendance_list:
    if student in attendance_count:
        attendance_count[student] += 1
    else:
        attendance_count[student] = 1

# Step 4: Printing the dictionary
print("Attendance Count Dictionary:", attendance_count)
```
